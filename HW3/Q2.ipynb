```python
# Import Your Libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Part 1

### Your Histogram Equalization Function

```python
def hist_equ(image):
    '''
    input:
    image (ndarray): input image
    output:
    output_image (ndarray): enhanced image
    '''
    
    ############
    # Your code
    # Start
    histogram = np.zeros((256), np.int)
    image = image.astype(int)
    np.add.at(histogram, image, 1)
    h = histogram
    out_image = image.copy()
    n_pixel = image.shape[0] * image.shape[1]
    transition = np.round(255 * h.cumsum() / n_pixel)
    output_image = transition[out_image]
    # End
    
    return output_image
```

```python
img = cv2.imread('River.jpg', 0)

### YOUR CODE ###
# START
equ = hist_equ(img)
# END

res = np.hstack((img, equ)) #stacking images side-by-side

plt.figure(figsize=(16, 16))
plt.imshow(res, cmap='gray')
```

### Histogram Equalization OpenCV Library

```python
img = cv2.imread('River.jpg', 0)

### YOUR CODE ###
# START
equ = cv2.equalizeHist(img)
# END

res = np.hstack((img, equ)) #stacking images side-by-side

plt.figure(figsize=(16, 16))
plt.imshow(res, cmap='gray')
```

## Part 2

### CLAHE OpenCV Library

```python
def CLAHE(image, gridSize, clip_limit):
    '''
    inputs:
    image (ndarray): input image
    gridSize (tuple): window size
    clip_limit (int): threshold for contrast limiting
    output:
    output_image (ndarray): improved image
    '''

    ############
    # Your code
    # Start
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=gridSize)
    
    # Applying the CLAHE algorithm to the image using the CLAHE object
    output_image = clahe.apply(image)
    # End

    return output_image
```

```python
img = cv2.imread('River.jpg', 0)

### YOUR CODE ###
# START
clh = CLAHE(img, (10,10), 2)
# END

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(14, 14),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(img,cmap='gray')
ax1.set_title('Source')
ax2.imshow(equ,cmap='gray')
ax2.set_title('Histogram Equalization')
ax3.imshow(clh,cmap='gray')
ax3.set_title('CLAHE')

plt.tight_layout()
plt.show()
```
