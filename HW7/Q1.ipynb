# Panorama with OpenCV

```python
import numpy as np
import cv2
import glob
import matplotlib.pylab as plt
```

```python
image_paths=['images/victoria1.png','images/victoria2.png','images/victoria3.png','images/victoria4.png','images/victoria5.png','images/victoria6.png','images/victoria7.png']

imgs = []
plt.figure(figsize=(6.4*7, 4.8*7), constrained_layout=False)
for i in range(len(image_paths)):
    imgs.append(cv2.imread(image_paths[i]))
    imgs[i]=cv2.resize(imgs[i],(0,0),fx=0.4,fy=0.4)
    plt.subplot(171+i), plt.imshow(imgs[i])

plt.show()
```

```python
stitchy=cv2.Stitcher.create()
(_,output)=stitchy.stitch(imgs)
```

```python
plt.figure(figsize=(20, 15), constrained_layout=False)
plt.imshow(output)
```
